```python
# Write a Python function that checks whether a passed string is 
# palindrome or not

def is_palindrome(s):
    """ Check if a string is a palindrome."""
    s = s.lower().replace(" ", "")
    return s == s[::-1]

word1 = "madam"
word2 = "hello"

print(f"'{word1}' is palindrome? {is_palindrome(word1)}")
print(f"'{word2}' is palindrome? {is_palindrome(word2)}")
```
