```python
# Write a python program to sum of three given integers. However if two values are equal sum will be be zero.

a = int(input("Enter the number 1: "))
b = int(input("Enter the number 2: "))
c = int(input("Enter the number 3: "))

if a == b or b == c or c == a:
    total = 0
else:
    total = a + b + c

print("Sum = ", total)
```
