```python
#  Write a Python function to insert a string in the middle of a string.

def insert_middle(str1, str2):
    middle = len(str1) // 2
    result = str1[:middle] + str2 + str1[middle:]
    return result

string1 = input("Enter the main string: ")
string2 = input("Enter the string to insert: ")

print("Result: ", insert_middle(string1, string2))
```
