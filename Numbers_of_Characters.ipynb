```python
# Write a python program to count the number of characters (character frequency) in a string.

string = input("Enter the string: ")
freq = {}

for char in string:
    if char in freq:
        freq[char] = freq[char] + 1
    else:
        freq[char] = 1

print("Character frequency")
for char in freq:
    print(char, ":", freq[char])
```
