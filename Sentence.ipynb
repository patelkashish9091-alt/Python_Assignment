```python
# Write a Python program to count the occurrences of each word in a given sentence.

sentence = input("Enter the sentence: ")
words = sentence.split()
count = {}

for word in words:
    if word in count:
        count[word] = count[word] + 1
    else:
        count[word] = 1
        
print("Word occurrences")

for word in count:
    print(word, ":", count[word])
```
