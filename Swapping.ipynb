```python
# Write a python program that swap two numbers with temp variable.

a = int(input("Enter the number 1: "))
b = int(input("Enter the number 2: "))

temp = a
a = b
b = temp

print(f"After Swapping: a = {a}, b = {b}")
```

```python
# Write a python program that swap two numbers without temp variable.

a = int(input("Enter the number 1: "))
b = int(input("Enter the number 2: "))

a, b = b, a
print(f"After Swapping: a = {a}, b = {b}")
```
