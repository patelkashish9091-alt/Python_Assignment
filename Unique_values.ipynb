```python
#  Write a Python program to get unique values from a list.

sample_list = [1, 2, 3, 2, 4, 1, 5, 3]

unique_values = list(set(sample_list))
print("Unique values (unordered): ", unique_values)

unique_ordered = []
for item in sample_list:
    if item not in unique_ordered:
        unique_ordered.append(item)

print("Unique values (ordered): ", unique_ordered)
```
